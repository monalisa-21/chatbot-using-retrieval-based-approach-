# code for classifier

```
# Step 1: Load the CSV file
df = pd.read_csv('/content/shuffled_modified4_file.csv', encoding='latin-1')

# Step 2: Separate features and labels
features = df['Question'].values.tolist()
labels = df['Domain'].values.tolist()

# # Step *: Preprocess labels
# labels = [label.strip() for label in labels]

# Step 3: Split the data into training and testing datasets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# Step 4: Feature Extraction
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)

# Step 5: Train a Classifier
classifier = LogisticRegression(max_iter=1000)
classifier.fit(X_train, y_train)

# Step 6: Model Evaluation
y_pred = classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
# print(classification_report(y_test, y_pred))

# Step 7: Predict and Classify
new_text = ["what are the benefits of Reliagrid digital solutions? "]
new_text_features = vectorizer.transform(new_text)
predicted_class = classifier.predict(new_text_features)
print("Predicted class:", predicted_class)
```
